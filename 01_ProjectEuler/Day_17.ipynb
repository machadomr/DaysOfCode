## <span style="color: blue;">Problem #29 </span>

### Consider all integer combinations of $a^b$ for $2 ≤ a ≤ 5$ and $2 ≤ b ≤ 5$:

### $2^2=4, 2^3=8, 2^4=16, 2^5=32$
### $3^2=9, 3^3=27, 3^4=81, 3^5=243$
### $4^2=16, 4^3=64, 4^4=256, 4^5=1024$
### $5^2=25, 5^3=125, 5^4=625, 5^5=3125$

### If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

### 4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125

### How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?

```python
def counting_distinct_terms():
    list_seen = set(i**j for i in range(2, 101) for j in range(2, 101))
    return str(len(list_seen))

print(counting_distinct_terms())
```

## <span style="color: blue;">Problem #30 </span>

### Surprisingly there are only three numbers that can be written as the sum of fourth powers of their digits:

### $1634 = 1^4 + 6^4 + 3^4 + 4^4$
### $8208 = 8^4 + 2^4 + 0^4 + 8^4$
### $9474 = 9^4 + 4^4 + 7^4 + 4^4$

### As $1 = 1^4$ is not a sum it is not included.

### The sum of these numbers is $1634 + 8208 + 9474 = 19316$.

### Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.

```python
# According to the problem, 1 = 1^5 is excluded !

def sum_calculation():
    output = sum(i for i in range(2, 1000000) if i == power_5_summation_digits(i))
    return str(output)

def power_5_summation_digits(n):
    return sum(int(j)**5 for j in str(n))

print(sum_calculation())
```
