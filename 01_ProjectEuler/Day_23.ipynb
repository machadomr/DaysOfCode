## <span style="color: blue;">Problem #41 </span>

### We shall say that an n-digit number is pandigital if it makes use of all the digits 1 to n exactly once. For example, 2143 is a 4-digit pandigital and is also prime.

### What is the largest n-digit pandigital prime that exists?

```python
import itertools

def checking_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**(0.5))+1):
        if num % i == 0:
            return False
    return True

def permutations(num):
    permutation_output = list(map("".join, itertools.permutations(str(num))))
    permutation_output = [int(i) for i in permutation_output]
    return permutation_output


# These are all the possibilities for rotations  (possib_rotations)
possib_rotations = [1, 12, 123, 1234, 12345, 123456, 1234567, 12345678, 123456789][::-1]
output = 0

for i in possib_rotations:
    for j in permutations(i):
        if checking_prime(j):
            output = max(output, j)

print(output)
```

## <span style="color: blue;">Problem #44 </span>

### Pentagonal numbers are generated by the formula, P_n=n(3n−1)/2. The first ten pentagonal numbers are:

### 1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

### It can be seen that P_4 + P_7 = 22 + 70 = 92 = P_8. However, their difference, 70 − 22 = 48, is not pentagonal.

### Find the pair of pentagonal numbers, P_j and P_k, for which their sum and difference are pentagonal and D = |P_k − P_j| is minimised; what is the value of D?

```python
def pentagonal_numbers():
    penta_list = set()
    num = 0
    while True:
        num += 1
        new_penta = num*(3*num - 1)/2
        penta_list.add(new_penta)
        for old_penta in penta_list:
            if new_penta-old_penta in penta_list and new_penta-2*old_penta in penta_list: 
                return new_penta-2*old_penta
print(pentagonal_numbers())
```
