## <span style="color: blue;">Problem #45 </span>

<p>Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:</p>
<table><tr><td>Triangle</td>
<td> </td>
<td>T<sub><i>n</i></sub>=<i>n</i>(<i>n</i>+1)/2</td>
<td> </td>
<td>1, 3, 6, 10, 15, ...</td>
</tr><tr><td>Pentagonal</td>
<td> </td>
<td>P<sub><i>n</i></sub>=<i>n</i>(3<i>n</i>−1)/2</td>
<td> </td>
<td>1, 5, 12, 22, 35, ...</td>
</tr><tr><td>Hexagonal</td>
<td> </td>
<td>H<sub><i>n</i></sub>=<i>n</i>(2<i>n</i>−1)</td>
<td> </td>
<td>1, 6, 15, 28, 45, ...</td>
</tr></table><p>It can be verified that T<sub>285</sub> = P<sub>165</sub> = H<sub>143</sub> = 40755.</p>
<p>Find the next triangle number that is also pentagonal and hexagonal.</p>

```python
def triangle_pentagonal_hexagonal():
    a = 285 + 1
    b = 165 + 1
    c = 145 + 1
    while True:
        triangle = a * (a + 1) // 2
        pentagon = b * (b * 3 - 1) // 2
        hexagon  = c * (c * 2 - 1)
        lower = min(triangle, pentagon, hexagon)
        if lower == max(triangle, pentagon, hexagon):
            return str(triangle)
        if lower == triangle: a += 1
        if lower == pentagon: b += 1
        if lower == hexagon : c += 1
            
print(triangle_pentagonal_hexagonal())
```
