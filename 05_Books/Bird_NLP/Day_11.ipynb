### 2) Use the corpus module to explore austen-persuasion.txt. How many word tokens does this book have? How many word types?

```python
# Removing non-alphabetic characters
```

```python
austen = 'austen-persuasion.txt'
enclitics = ("d", "ll", "m", "re", "s", "t", "ve")
words = gutenberg.words(austen)
```

```python
number_tokens = len([w for w in words if w.isalpha() and w not in enclitics])
number_tokens
```

```python
# number of word types
number_words_types = len(set([w for w in words if w.isalpha() and w not in enclitics]))
number_words_types
```

```python
# Without the removal of non-alphabetic characters
```

```python
number_tokens = len([w for w in words])
number_tokens
```

```python
number_words_types = len(set([w for w in words]))
number_words_types
```

### 4) Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
count_freq = nltk.ConditionalFreqDist(
        (target, fileid[:4])
        for fileid in state_union.fileids()
        for w in state_union.words(fileid)
        for target in ['men', 'women', 'people']
        if w.lower().startswith(target))
```

```python
# We will have too many info -- visualization will be bad!
len(state_union.fileids())
```

```python
# We can sample one state of the Union every four years!
years = range(1945, 2007, 4)
count_freq.tabulate(samples = [str(y) for y in years], conditions = ['men', 'women', 'people'])
```

```python
# We can also see what we have in the table in a plot:
plt.figure(figsize=(15, 10))
count_freq.plot();
```
