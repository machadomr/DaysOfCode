### Web Scrapping - Using [Beautiful Soup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/)

```python
from pprint import pprint
from bs4 import BeautifulSoup
from urllib.request import urlopen 
```

```python
# Define webpage, make it html, then parse using beautiful soup
myurl = "https://stackoverflow.com/questions/2081586/web-scraping-with-python" 
html = urlopen(myurl).read()
soupified = BeautifulSoup(html, 'html.parser') 
```

```python
# Printing the html page
pprint(soupified.prettify())
```

```python
# Printing the title of the webpage
soupified.title
```

```python
# Presenting the question from the website and the best given answer
question = soupified.find("div", {"class": "question"}) 
questiontext = question.find("div", {"class": "s-prose js-post-body"})
print("Problem: \n", questiontext.get_text().strip())

answer = soupified.find("div", {"class": "answer"}) 
answertext = answer.find("div", {"class": "s-prose js-post-body"})
print("Good Solution: \n", answertext.get_text().strip())
```
