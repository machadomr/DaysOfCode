## DPT in Textual Features
### Cleaning, Stop Words, Tokenize, Steem, Rejoin

```python
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import random
```

```python
loans = pd.read_csv("data/text_df.csv",low_memory=False)
loans.shape
```

```python
main_df = loans.copy()
```

```python
main_df
```

```python
pd.options.display.max_colwidth = 300
text_df = main_df[['processed']]
text_df
```

```python
# Load the regular expression library
text_df = text_df.replace(np.nan, '', regex=True)

import re
# Remove punctuation
text_df['processed'] = \
text_df['processed'].map(lambda x: re.sub('[,\.!?></]', '', x))
# Convert the titles to lowercase
text_df['processed'] = \
text_df['processed'].map(lambda x: x.lower())
# Print out the first rows of papers
text_df
```

```python
text_df['processed'] = text_df['processed'].replace('0','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('1','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('2','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('3','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('4','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('5','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('6','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('7','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('8','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('9','',regex=True).astype(str)
text_df['processed'] = text_df['processed'].replace('borrower added on ','',regex=True).astype(str)
```

```python
import nltk

def identify_tokens(row):
    desc = row['processed']
    tokens = nltk.word_tokenize(desc)
    # taken only words (not punctuation)
    token_words = [w for w in tokens if w.isalpha()]
    return token_words

text_df['words'] = text_df.apply(identify_tokens, axis=1)
text_df
```

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()

my_list = []
```

```python
def stem_list(row):
    my_list = row['words']
    stemmed_list = [stemming.stem(word) for word in my_list]
    return (stemmed_list)

text_df['stemmed_words'] = text_df.apply(stem_list, axis=1)
text_df
```

```python
from nltk.corpus import stopwords

stops = stopwords.words("english")
# to add words to the stop words list!!!
stops.append('THIS_WORD_CAN_BE_ADDED')

def remove_stops(row):
    my_list = row['stemmed_words']
    meaningful_words = [w for w in my_list if not w in stops]
    return (meaningful_words)

text_df['stem_meaningful'] = text_df.apply(remove_stops, axis=1)
text_df
```

```python
def rejoin_words(row):
    my_list = row['stem_meaningful']
    joined_words = ( " ".join(my_list))
    return joined_words

text_df['processed_done'] = text_df.apply(rejoin_words, axis=1)
text_df
```

```python
cols_to_drop = ['processed', 'words', 'stemmed_words', 'stem_meaningful']
text_df.drop(cols_to_drop, inplace=True, axis=1)
text_df
```
