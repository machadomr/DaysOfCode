```python
# Import of different libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

# Pandas Settings
pd.set_option('display.max_columns', 100)
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#plt.figure(figsize=(25,30)) 

labels = ['LDA', 'NMF', 'LSA', 'Top2Vec', 'BERT']
father_means = [5, 5, 5, 5, 5]
mother_means = [3, 3, 3, 3, 3]
son_means = [4, 4, 4, 4, 4]
daughter_means = [2, 2, 2, 2, 2]
a_means = [5, 5, 5, 5, 5]
b_means = [3, 3, 3, 3, 3]
c_means = [4, 4, 4, 4, 4]
d_means = [2, 2, 2, 2, 2]

num_items = len(labels)
x = np.arange(num_items)  # the label locations
width = 0.35  # the width of the bars

margin = 0.05
width = (1.-2.*margin)/num_items

#fig, ax = plt.subplots(figsize=(20, 10))

fig, ax = plt.subplots(figsize=(20, 15))
rects1 = ax.bar(x - width/2, father_means, width, label='Father', color="gray")
rects1 = ax.bar(x - width/2, mother_means, width, bottom=father_means, label='Mother')
rects2 = ax.bar(x + width/2, son_means, width, label='Son')
rects2 = ax.bar(x + width/2, daughter_means, width, bottom=son_means, label='Daughter')
rects3 = ax.bar(x + 1.5*width, a_means, width, label='a_means')
rects3 = ax.bar(x + 1.5*width, b_means, width, bottom=b_means, label='b_means')
rects3 = ax.bar(x - 1.5*width, c_means, width, label='c_means')
rects3 = ax.bar(x - 1.5*width, d_means, width, bottom=d_means, label='d_means')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
#ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()



plt.show()
```
